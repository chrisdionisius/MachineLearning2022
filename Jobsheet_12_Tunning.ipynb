```python
# baca data
data = pd.read_csv('/content/drive/MyDrive/dataset/wbc.csv')
data.head()
```

```python
# slice kolom terakhir --> Unnamed column
data = data.iloc[:,:-1]

data.head()
```

```python
X = data.iloc[:, 2:]
y = data['diagnosis']

X.head()
```

```python
from sklearn.model_selection import RepeatedKFold

# inisiasi repated k-fold
cv = RepeatedKFold(n_splits=4, n_repeats=3, random_state=42)
```

```python
# from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble  import RandomForestClassifier
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import uniform, truncnorm, randint

# inisiasi model
dt = RandomForestClassifier()

# Definisikan hyperparameter yang akan digunakan
params = {
    # randomly sample numbers from 4 to 204 estimators
    'n_estimators': randint(4,200),
    # normally distributed max_features, with mean .25 stddev 0.1, bounded between 0 and 1
    'max_features': truncnorm(a=0, b=1, loc=0.25, scale=0.1),
    # uniform distribution from 0.01 to 0.2 (0.01 + 0.199)
    'min_samples_split': uniform(0.01, 0.199)
}

# inisiasi randomized search berdasarkan nilai repeated k-fold dan hyperparameter
rsc = RandomizedSearchCV(dt,params, cv=cv, n_iter=100, random_state=1)

# train the random search meta-estimator to find the best model out of 100 candidates
model = rsc.fit(X, y)

# Evaluasi dengan score
score = model.score(X,y)

print(f'Hasil evaluasi: {score}')
print(f'Konfigurasi hyperparameter: {rsc.best_params_}')
```

```python
# import support vector classifier 
# "Support Vector Classifier"
from sklearn.svm import SVC

# inisiasi model
svclassifier = SVC(kernel='linear')

model = svclassifier.fit(X, y) 
print('w = ',svclassifier.coef_)
print('b = ',svclassifier.intercept_)
print('Indices of support vectors = ', svclassifier.support_)
print('Support vectors = ', svclassifier.support_vectors_)
print('Number of support vectors for each class = ', svclassifier.n_support_)
print('Coefficients of the support vector in the decision function = ', np.abs(svclassifier.dual_coef_))

# Evaluasi dengan score
score = model.score(X,y)

print(f'Hasil evaluasi: {score}')
print(f'Konfigurasi hyperparameter: {rsc.best_params_}')
```

```python
# inisiasi model
svclassifier = SVC(kernel='poly', degree=8)

model = svclassifier.fit(X, y)

# Evaluasi dengan score
score = model.score(X,y)

print(f'Hasil evaluasi: {score}')
print(f'Konfigurasi hyperparameter: {rsc.best_params_}')
```
